```python

def at_least_two_elements_similar(list_of_lists):
    # Iterate through pairs of inner lists
    for i in range(len(list_of_lists)):
        for j in range(i + 1, len(list_of_lists)):
            count_similar = sum(1 for x, y in zip(list_of_lists[i], list_of_lists[j]) if x == y)
            if count_similar >= 2:
                return True
    return False
at_least_two_elements_similar(song_name_genre)
```

```python
import os

def remove_txt_files_from_directories(directory_list):
    for directory in directory_list:
        # Check if the directory exists
        if os.path.exists(directory):
            # Get a list of all files in the directory
            files = os.listdir(directory)
            # Iterate through the files
            for file in files:
                # Check if the file ends with '.txt'
                if file.endswith('.txt'):
                    file_path = os.path.join(directory, file)
                    # Remove the .txt file
                    os.remove(file_path)
                    print(f"Removed: {file} from {directory}")
        else:
            print(f"The directory '{directory}' does not exist.")

# Example usage:
directory_list = listf
remove_txt_files_from_directories(directory_list)
```
